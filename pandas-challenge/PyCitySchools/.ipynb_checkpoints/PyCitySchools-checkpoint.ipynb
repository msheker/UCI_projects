```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
results = dict()
schools = df.school_name.value_counts()

results["Total Schools"] = [len(schools)]
results["Total Students"] = [schools.sum()]
results["Total Budget"] = [f"${df.groupby('school_name').mean().budget.sum()}"]

results["Average Math"] = [df.math_score.mean()]
results["Average Reading"] = [df.reading_score.mean()]

rdf = df.reading_score >= 70 
mdf = df.math_score >= 70

results["Passing Math"] = [len(df.loc[df.math_score >= 70,"student_name"])*100/schools.sum()]
results["Passing Reading"] = [len(df.loc[df.reading_score >= 70,"student_name"])*100/schools.sum()]
results["Passing Both"] = [(mdf & rdf).value_counts()[True]*100/schools.sum()]

results = pd.DataFrame(results)
results
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
results = dict()
avg_df = df.groupby("school_name").mean()
avg_df = avg_df.loc[:,["reading_score","math_score"]]

for school in df.school_name.unique():
    cur = df.loc[df.school_name == school,:]
    #print(cur)
    results[school] = [cur.iloc[1,8],cur.iloc[1,9],cur.iloc[1,10],cur.iloc[1,10]/cur.iloc[1,9]]
    results[school].append(avg_df.loc[school,["math_score"]].iloc[0])
    results[school].append(avg_df.loc[school,["reading_score"]].iloc[0])
    
    crdf = cur.reading_score >= 70 
    cmdf = cur.math_score >= 70
    
    results[school].append(len(cur.loc[cur.math_score >= 70,"student_name"])*100/cur.iloc[1,9])
    results[school].append(len(cur.loc[cur.reading_score >= 70,"student_name"])*100/cur.iloc[1,9])
    results[school].append((cmdf & crdf).value_counts()[True]*100/cur.iloc[1,9])
    
sdf = pd.DataFrame(results).T.sort_index()
sdf.columns = ["School Type","Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]
sdf
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spdf = sdf
spdf["Spending Ranges (Per Student)"] = pd.cut(spdf["Per Student Budget"],[0,585,630,645,675],labels=["<$585","$585-629","$630-644","$645-675"])
spdf = spdf.loc[:,["Spending Ranges (Per Student)",'Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', '% Overall Passing',]]
spdf = spdf.groupby(["Spending Ranges (Per Student)"]).max()
spdf
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
scdf = sdf
scdf["School Size"] = pd.cut(scdf["Total Students"],[0,1000,2000,5000],labels=["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"])
scdf = scdf.loc[:,["School Size",'Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', '% Overall Passing',]]
scdf = scdf.groupby(["School Size"]).max()
scdf
```
